## GROUP BY, COUNT, ORDER BY

```python
"""
select iso_country, type, count(*) 
from airports 
group by iso_country, type 
order by iso_country, type
"""
airports.groupby(['iso_country','type']).size().to_frame('size')
#size() includes NaN values, count() does not
```

```python
"""
select iso_country, type, count(*) 
from airports 
group by iso_country, type 
order by iso_country, count(*) desc
"""
airports.groupby(['iso_country','type']).size().to_frame('size').sort_values(['iso_country','size'],ascending=[True,False])
```

```python
#What is this trickery with .to_frame() and .reset_index()? Because we want to sort by our calculated field (size), this field needs to become part of the DataFrame. After grouping in Pandas, we get back a different type, called a GroupByObject. So we need to convert it back to a DataFrame. With .reset_index(), we restart row numbering for our data frame.
```

## Having

```python
"""
select type, count(*) 
from airports 
where iso_country = 'US' 
group by type 
having count(*) > 1000 
order by count(*) desc
"""
airports[airports.iso_country == 'US'].groupby('type').filter(lambda g: len(g) > 1000).groupby('type').size().sort_values(ascending=False)
```

## Window Function

### Cumulative Sum

```python
df = (df.assign(total=df.sort_values(['ts'])
                                .groupby(['url', 'service'])
                                .delta
                                .cumsum()))
"""
select 
    *, 
    sum() over(partition by url, service order by ts) rank
from df
"""
```

### Lag

```python
df = (df.assign(delta=df.sort_values(['ts'])
                                .groupby(['url', 'service'])
                                .diff(1)))
df.loc[df['delta'].isnull(), 'delta'] = df['total']
"""
select
    *,
     lag(total,1) over(partition by url, service order by ts)
"""
```

### Rank

```python
df = (df.assign(rn=df.groupby(['service'])['total']
                              .rank(method='first', ascending=False))
                              .query('rn <= 3')
                              .sort_values(['service', 'rn']))
"""
select *
    rank() over(partition by service order by service) rank
from df
"""
```

### Rolling Window

```python
# mean of current and previous 2 rows
ro = (df.groupby(['url', 'service'])
                 .rolling(3, on='ts', min_periods=1)
                 .mean()
                 .reset_index(drop=True))
# mean of last 5 mins
# ro = (df.groupby(['url', 'service'])
#                  .rolling('5min', on='ts', min_periods=1)
#                  .mean()
#                  .reset_index(drop=True))

df = df.merge(ro,
                       on=['url', 'service', 'ts'],
                       suffixes=('', '_avg3'))
```
